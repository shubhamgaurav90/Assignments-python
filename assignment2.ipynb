```python
#write a user defined function that takes a number as input and returns positive even, positive odd, negative, odd
def check_num(num):
    if num>0 and num%2==0:  #for positive and even
        print(f"{num} is positive even number") 
    elif num<0:    #for negative
        print(f"{num} is negative number")
    elif num>0 and num%2!=0:    #for positive and odd
        print(f"{num} is positive odd number")
    elif num==0:
        print(f"{num} is zero")
n=int(input("enter the number:- "))
check_num(n)
    
    
    
    
```

```python
#question 2:
def compare_int(a,b,c):
    if a>b and a>c:
        print(f"{a} is greatest among {a},{b},{c}")
    elif b>a and b>c:
        print(f"{b} is greatest among {a}, {b}, {c}")
    elif c>a and c>b:
        print(f"{c} is greatest among {a},{b},{c}")
    else: 
        print(f"{a},{b},{c} are equal")
a=int(input("enter the first number:- "))
b=int(input("enter the second number:- "))
c=int(input("enter the third number:- "))
compare_int(a,b,c)
```

```python
#question 3:
def result(num):
    if num>=90:
        print("A grade ")
    elif num>=75 and num<=89:
        print("B grade")
    elif num>=50 and num<=74:
        print("C grade")
    elif num<50:
        print("Failed")
n=int(input("enter the marks:- "))
result(n)
```

```python
#question 4:
def classify_age(age):
    if age<0:
        raise ValueError("Age cannot be negative")
    elif age<18:
        print("MINOR")
    elif age>=18:
        print("ADULT")
N=int(input("enter the age:- "))
try:
    classify_age(N)
except ValueError as e:
    print(f"Error: {e}")
```

```python
def count_even_numbers(nums):
    even_count = 0  #behaves liek a counter started from zero
    
    for num in nums:
        if num == 0:    # Stop processing if zero appears
            break
    
        if num < 0: # Ignore negative numbers
            continue
        if num % 2 == 0: #checks for the even number if any and increases the count by 1
            even_count += 1
       
    return even_count
print(count_even_numbers([2, 4, -3, 7, 8, 0, 10])) ##this will directly prints the total even nos. out of this list

```
