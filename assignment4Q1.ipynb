```python
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

```python
df=sns.load_dataset('iris')
print(df.shape)
```

```python
print(df.isnull().sum())
```

```python
df.describe()
```

```python
df_filtered=df[df["species"].isin(["versicolor","virginica"])]
df_final=df_filtered[df_filtered["petal_length"]>df['petal_length'].mean()]
```

```python
x_train,x_test,y_train,y_test=train_test_split(df_final.drop('species',axis=1),df_final['species'],test_size=0.2,random_state=42)
```

```python
model=SVC(kernel='linear').fit(x_train,y_train)
```

```python
accuracy=accuracy_score(y_test,model.predict(x_test))
print(accuracy)
```

```python
classification_report(y_test,model.predict(x_test))
```
